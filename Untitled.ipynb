```python
def read_json(json_file):
    with open(json_file, 'r') as myfile:
        data=myfile.read()
    arr = json.loads(data)
    return arr
```

```python
gun_laws = read_json('json/state_gun_laws.json')
```

```python
gun_laws_df = pd.read_json('json/state_gun_laws.json').transpose()
gun_laws_df
```

```python
row = {'Long Guns': 1.0, 'Handguns': 1.0}
row['Long Guns'] + row['Handguns']

def parse_row(row):
    if row == 0:
        return row
    else:
        if row['Long Guns'] + row['Handguns'] == 0:
            return 0
        else:
            return 1
        
    
    
```

```python
assault_weapon_laws = gun_laws_df['Assault weapon law?']
assault_weapon_laws.fillna(0, inplace=True)
assault_weapon_law_df = assault_weapon_laws.apply(parse_row).to_frame()
assault_weapon_law_df = assault_weapon_law_df.reset_index()
assault_weapon_law_df = assault_weapon_law_df.rename(columns={"index":"State", "Assault weapon law?": "assault_weapon_law"})
assault_weapon_law_df = assault_weapon_law_df.append({'State': 'WY', 'assault_weapon_law':0}, ignore_index = True)
assault_weapon_law_df
```

```python
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
state_geo = f'{url}/us-states.json'
state_unemployment = f'{url}/US_Unemployment_Oct2012.csv'
state_data = pd.read_csv(state_unemployment)

m = folium.Map(location=[48, -102], zoom_start=3)

folium.Choropleth(
    geo_data=state_geo,
    name='choropleth',
    data=assault_weapon_law_df,
    columns=['State', 'assault_weapon_law'],
    key_on='feature.id',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Assault Weapon Law'
).add_to(m)

folium.LayerControl().add_to(m)


colors = {True: 'red', False: 'blue'}

data.apply(lambda row:folium.CircleMarker(location=[row["latitude"], row["longitude"]], 
                                              radius=(row['fatalities']/data['fatalities'].max())*30, color = colors[row['assault_weapon']], fill_color=colors[row['assault_weapon']], popup = row['summary'])
                                             .add_to(m), axis=1)


m.add_child(folium.map.LayerControl())

m
# m.save('assault_weapon_laws.html')
```
